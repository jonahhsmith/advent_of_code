```python
d2 = aocd.get_data(session=session_id,day=2,year=2015)
```

```python
lines = d2.split()
```

```python
sf = 0
for line in lines:
    a,b,c = [int(x) for x in line.split('x')]
    s1 = a*b
    s2 = a*c
    s3 = b*c
    area = 2* (s1 + s2 + s3) + min(s1, s2, s3)
    sf += area

print(sf)
```

```python
r = 0
for line in lines:
    a,b,c = [int(x) for x in line.split('x')]
    rl = 2 * (a + b + c - max(a,b,c)) + a*b*c
    r += rl

print(r)
```

```python
from collections import defaultdict
import operator

def read_graph(data,rep=None,val=None):
    graph = defaultdict(list)
    for line in data:
        split = line.split()
        if len(split) == 3:
            graph[split[-1]] = ("EQ", split[0])
        elif len(split) == 4:
            graph[split[-1]] = (split[0], split[1])
        else:
            graph[split[-1]] = (split[1], split[0], split[2])
    if rep:
        graph[rep] = val
    return graph

def op_eq(value):
    return value

def op_not(value):
    return ~value & 0xffff

OPERATIONS = {"EQ": op_eq,
              "NOT": op_not,
              "AND": operator.iand,
              "OR": operator.ior,
              "RSHIFT": operator.rshift,
              "LSHIFT": operator.lshift}

def memoize(f):
    memo = {}
    print('memoize:',f)
    def helper(graph, key):
        # print(graph['b'])
        print('helper:',key, memo.get(key,'None'))
        if key not in memo:
            memo[key] = f(graph, key)
        return memo[key]
    return helper

@memoize
def find_key(graph, key):
    try:
        return int(key)
    except ValueError:
        pass
    print('find_key:', key, graph.get(key))
    value = graph[key]
    op = value[0]

    if len(value) == 2:
        return OPERATIONS[op](find_key(graph, value[1]))
    else:
        return OPERATIONS[op](find_key(graph, value[1]), find_key(graph, value[2]))

def main():
    # graph = read_graph(inst)
    # print(graph['b'])
    # p1 = find_key(graph, 'a')
    # print(p1)

    graph2 = read_graph(inst,'b',('EQ','3176'))
    print(find_key(graph2, 'a'))


if __name__ == "__main__":
    main()
```
